```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(numerical)
numerical_scaled = scaler.transform(numerical)
cols = scaler.get_feature_names_out(input_features=numerical.columns)
numerical_scaled = pd.DataFrame(numerical_scaled ,columns=cols)
numerical_scaled
# scaling before kbest
```

```python
X = numerical_scaled
y = targets['TARGET_B']
num_for_k = 39 # why 39? see a few cells below
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
model = SelectKBest(chi2, k=num_for_k).fit(X, y) # fitting the model, just to check the scores
df = pd.DataFrame(data = model.scores_, columns = ['score']) # displaying scores as dataframe
df['column'] = numerical.columns
print(df.sort_values(by = ['score'], ascending = False).head(num_for_k))
# columns printed below will be columns that i want to take
```

```python
len(df.column[df.score >= df.score.mean()])
# this is where the number 39 came from. I picked the columns, which scores are
# above or equals the average scores
```

```python
cols = df.sort_values(by = ['score'], ascending = False).head(num_for_k).iloc[:,1].tolist()
cols
# to get 39 colummn names with highest scores. sorted descending
```

```python
X_selected = numerical_scaled.loc[:,cols] # get only those 39 columns
display(X_selected.shape)
X_selected.head()
```

```python
X_corr = pd.concat([targets.TARGET_B, X_selected], axis = 1).corr()
# i want to check whether or not the 39 columns have at least a decent
# collinearity score with our target B, also to identify multicollinearity
negative_corr = X_corr.TARGET_B[X_corr.TARGET_B < 0].index.values.tolist()
# remove columns which have negative collinearity with target B
high_corr = X_corr.TARGET_B[X_corr.TARGET_B > 0.87].index.values.tolist()
# remove columns which have too high of collinearity with target B
to_drop = negative_corr + high_corr # collect the names i want to drop
print(len(X_selected.columns))
for i in to_drop:
    try: # i like using try except. feels safer that way
        X_selected = X_selected.drop(columns = i)
    except:
        pass
X_corr = pd.concat([targets.TARGET_B, X_selected], axis = 1).corr()
# make a new X_corr because the target B was deleted. target B has collinearity of 1 with target B.
print(len(X_selected.columns))
# first num: columns before dropping
# second num: columns after dropping
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax = sns.heatmap(X_corr, annot=True)
plt.show()
```
